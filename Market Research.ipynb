```python
# (6)   Hourly performance.  We are very interested to learn which hour of the day is the best time to send an email.   
#(6.a) Please calculate by hour of the day, total emails sent, opened and clicked.  Totals and percentages. 
df['Message Date\n(first msg)']=pd.to_datetime(df['Message Date\n(first msg)']) 
```

```python
sent=df_data_1
sent=sent[['Date','Status']]
sent.set_index('Date',inplace=True)
z=sent.groupby(sent.index.hour).count()
z['Percent']= (z['Status'] / z['Status'].sum()) * 100
z.head(10)
```

```python

opens=df_data_1[['Date','Opens']]
opens.set_index('Date',inplace=True)
x=opens.groupby(opens.index.hour).sum()
x['Percent']= (x['Opens'] / x['Opens'].sum()) * 100
x.head(20)
```

```python
click=df_data_1[['Date','Clicks']]
click.set_index('Date',inplace=True)
y=click.groupby(click.index.hour).sum()
y['Percent']= (y['Clicks'] / y['Clicks'].sum()) * 100
y.head(10)
```

```python
r=z
r['Opens']=x['Opens']
r['Percent opened']=x['Percent']
r['Clicks']=y['Clicks']
r['Percent clicked']=y['Percent']
r=r.rename(columns={'Status':'Sent','Percent':'Persent sent'})
r.index.names = ['Hour of day']
r.head(10)
```
